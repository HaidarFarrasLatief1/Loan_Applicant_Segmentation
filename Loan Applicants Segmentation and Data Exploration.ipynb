### 4.2 Define Functions

### 4.2.1 For observation

#### 4.2.1.1 plot_data_distribution

This function are used in 4.5.1.2.2 'Missing Value Unique Value & Data Distribution' for numerical features.

````python
def plot_data_distribution(data):
    plt.hist(data, bins='auto', alpha=0.7, rwidth=0.85)
    plt.grid(axis='y', alpha=0.5)
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Data Distribution')
    plt.show()
````

#### 4.2.1.2 get_formatted_value_counts

This function are used in 4.5.1.2.1 'Missing Value Dataframe' and 4.5.2 'Unique Value' for data obseration.

````python
def get_formatted_value_counts(data_series):
    value_counts = data_series.value_counts()
    total_count = len(data_series)

    percentage_counts = (value_counts / total_count) * 100
    formatted_counts = percentage_counts.apply(lambda x: f"{x:.2f}%")

    result = pd.concat([value_counts, formatted_counts], axis=1)
    result.columns = ['Count', 'Percentage']

    result = result.sort_index()

    return result
````

### 4.2.2 For handling outlier

#### 4.2.2.1 find_skewed_boundaries

This function are used in 4.5.3.1 'Outlier Observation' for data outlier observation.

````python
def find_skewed_boundaries(df, variable, distance):
    lower_quantile = df[variable].quantile(0.25)
    upper_quantile = df[variable].quantile(0.75)
    IQR = upper_quantile - lower_quantile

    lower_boundary = lower_quantile - (IQR * distance)
    upper_boundary = upper_quantile + (IQR * distance)

    return lower_boundary, upper_boundary
````

#### 4.2.2.2 diagnostic_plots

This function are used in 4.5.3.1 'Outlier Observation' for data outlier observation.

````python
def diagnostic_plots(data, feature):
    plt.figure(figsize=(16, 4))

    plt.subplot(1, 3, 1)
    sns.histplot(data[feature], kde=True)
    plt.title('Histogram with KDE')

    plt.subplot(1, 3, 2)
    stats.probplot(data[feature], dist="norm", plot=plt)
    plt.title('Q-Q Plot')

    plt.subplot(1, 3, 3)
    sns.boxplot(y=data[feature])
    plt.title('Boxplot')

    plt.tight_layout()
    plt.show()
````

#### 4.2.2.3 remove_outliers

This function are used in 4.5.3.2 'Removing outliers' to remove outlier data.

````python
def remove_outliers(data, variable, lower_bound, upper_bound):
    data_without_outliers = data[(data[variable] >= lower_bound) & (data[variable] <= upper_bound)]
    return data_without_outliers
````

### 4.2.3 For exploration data analysis

#### 4.2.3.1 plot_Loan_Status_pie

This function are used in 4.6.1.1 'Loan_Status segmentation' for categorical data observation in pie chart.

````python
def plot_Loan_Status_pie(data, column, Loan_Status_labels):
    num_Loan_Statuss = len(Loan_Status_labels)

    plt.figure(figsize=(5 * num_Loan_Statuss, 5))

    for i, label in enumerate(Loan_Status_labels):
        plt.subplot(1, num_Loan_Statuss, i + 1)
        segment_data = data[data['Loan_Status'] == label]
        segment_counts = segment_data[column].value_counts()

        # Create pie chart
        plt.pie(segment_counts, labels=segment_counts.index, autopct='%1.1f%%')

        plt.title('Loan_Status {}: {}'.format(label, column), size=16)

    # Set plot background color to white
    fig = plt.gcf()
    fig.set_facecolor('white')

    plt.tight_layout()
    plt.show()
````

### 4.2.4 For correlation data analysis

#### 4.2.4.1 plot_correlation for correlation between numerical features

This function used in 4.6.3.2 'Correlation' for correlation between numerical features.

````python
def plot_correlation(data, column1, column2):
    correlation = data[column1].corr(data[column2])

    plt.scatter(data[column1], data[column2])
    plt.xlabel(column1)
    plt.ylabel(column2)
    plt.title('Correlation: {:.2f}'.format(correlation))
    plt.show()

    return correlation
````

#### 4.2.4.2 categorical_numerical_correlation for correlation between categorical and numerical features

This function used in 4.6.4.2 'Correlation' for correlation between categorical and numerical features.

````python
def categorical_numerical_correlation(data, categorical_column, numerical_column):
    groups = data.groupby(categorical_column)[numerical_column].apply(list)
    result = stats.f_oneway(*groups)

    p_value = result.pvalue
    if p_value < 0.05:
        correlation_strength = 'significant'
    else:
        correlation_strength = 'not significant'

    correlation_explanation = "The correlation between {} and {} is {} (p-value: {:.4f}).".format(
        categorical_column, numerical_column, correlation_strength, p_value
    )
    
    return correlation_explanation
````

##### 4.6.2.2.1 Preparing the data

````python
df_clstr_prprcs
````

````python
cluster_data = df_clstr_prprcs[['Gender','Married','Dependents','Education','Self_Employed','ApplicantIncome','CoapplicantIncome','LoanAmount','Loan_Amount_Term','Credit_History','Property_Area','Loan_Status']]
````

````python
cluster_data.info()
````

##### 4.6.2.2.2 Model Evaluation

4.6.2.2.2.1 Elbow Method

````python
# Define a range of k values to evaluate
k_values = range(1, 10)
# Initialize a list to store the cost values
costs = []

# Display loading bar
with tqdm(total=len(k_values), desc="Progress") as pbar:
    for k in k_values:
        # Apply K-Prototypes clustering
        kproto = KPrototypes(n_clusters=k, init='Cao')
        kproto.fit(cluster_data.values, categorical=[0, 1, 2, 3, 4, 9, 10, 11])

        # Append the cost value to the list
        costs.append(kproto.cost_)

        # Update loading bar
        pbar.update(1)
````

````python
# Plot the elbow curve
plt.plot(k_values, costs, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Cost')
plt.title('Elbow Method')
plt.show()
````

4.6.2.2.2.2 Silhouette Method

````python
# Define range of k values to evaluate
k_values = range(2, 10)

# Initialize a list to store the silhouette scores
silhouette_scores = []

# Display loading bar
with tqdm(total=len(k_values), desc="Progress") as pbar:
    for k in k_values:
        # Apply K-Prototypes clustering
        kproto = KPrototypes(n_clusters=k, init='Cao')
        kproto.fit(cluster_data, categorical=[0, 1, 2, 3, 4, 9, 10, 11])

        # Assign cluster labels
        labels = kproto.labels_

        # Calculate silhouette score
        silhouette_avg = silhouette_score(cluster_data, labels)

        # Append the silhouette score to the list
        silhouette_scores.append(silhouette_avg)

        # Update loading bar
        pbar.update(1)
````

````python
# Plot the silhouette scores
plt.plot(k_values, silhouette_scores, 'bx-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Method')
plt.show()
````

4.6.2.2.2.3 Model evaluation conclusion
